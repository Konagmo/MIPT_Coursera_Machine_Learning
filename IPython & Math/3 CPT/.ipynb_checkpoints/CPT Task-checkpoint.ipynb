```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Выберем распределение хи-квадрат. Для определенности возьмем 4 степени свободы.

Создаем выборку от 0 до 100 объема 1000.

```python
x = np.linspace(0, 20, 100)
```

```python
n = 1000
k = 4
rv = sts.chi2(k)
sample = rv.rvs(n)
```

Сторим теоретическую плотность и гистограмму выборки на одном графике.

```python
plt.hist(sample, normed=True, label='histogram', alpha=0.5, bins=20)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')

pdf = rv.pdf(x)
plt.plot(x, pdf, label='theoretical density')
plt.legend()
```

Как видно, гистрограмма и теоретическое распределение почти совпадают.

Рассчет нормального распределения
- Дисперсия: $\sigma^2$ = 2k; Математическое ожидание: $\mu = k$ 
- По ЦПТ: $N(EX,\frac{DX}{n})$ - нормальное распределение, аппроксимирующее распределение выборочных средних ->  $ N(k, \frac{2k}{n})$

```python

def chi_func_task(n):
    means = []
    for i in range(1000):
        s = rv.rvs(n)
        means.append(np.mean(s)) 
    
    plt.hist(means, bins = 20, label='hist', normed = True, alpha=0.5)
    
    x = np.linspace(0, 20, 100)
    
    norm_rv = sts.norm(rv.mean(), np.sqrt(rv.var() / float(n)))  
    pdf = norm_rv.pdf(x) 
    
    plt.plot(x, pdf, label='theoretical pdf')
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    
    plt.legend()
    plt.show()
```

```python
for i in [5,10,50]:
    chi_func_task(i)
```

С ростом n точность аппроксимации увеличивается, что не удивительно, ведь увеличивая объем выборки, мы повышаем точность нашего результата.

При маленьких n графики шире, чем при больших, что тоже описывает лучшее усреднение величин при росте объема выборок.
